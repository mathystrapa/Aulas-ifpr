```python
# NESTA AULA, APRENDEREMOS A CRIAR UM FORMULÁRIO SIMPLES USANDO A BIBLIOTECA TKINTER. LEMBRE-SE DE EXECUTAR OS COMANDOS NO ARQUIVO 'main.py' PARA NÃO TRAVAR.

import tkinter as tk
from tkinter import *

def get_info():
    global nome,idade
    nome = var_nome.get()
    idade = var_idade.get()
    janela_formulario.quit()

janela_formulario = tk.Tk()     # CRIA A JANELA DO FORMULÁRIO

janela_formulario.geometry('400x600')   # DEFINE O TAMANHO DA JANELA PARA 400px DE LARGURA E 600px DE ALTURA

janela_formulario.title('Formulário de cadastro')   # COLOCA O TÍTULO DA JANELA

janela_formulario.configure(bg='#addef0')   # COLOCA A COR DE FUNDO

titulo = tk.Label(janela_formulario, text="Formulário teste", background='#addef0', font=('Helvetica', 20, 'bold'))    # CRIA O TÍTULO NA FONTE 'Helvetica', COM TAMANHO 20 E EM NEGRITO

titulo.pack(pady=10)     # INSERE O TÍTULO NA JANELA DO FORMULÁRIO

formulario=tk.Frame(janela_formulario, width=300, height=450, bg='#c9c9c9', bd=10, relief=tk.SUNKEN)
formulario.pack(pady=5)     # CRIA UM FRAME DENTRO DA JANELA

label_nome = tk.Label(formulario, text="Insira seu nome:", font=('Arial', 10), bg='#c9c9c9')    # CRIA UM RÓTULO PARA O NOME
label_nome.pack(pady=5)     # INSERE O RÓTULO

var_nome = tk.StringVar()   # CRIA UMA VARIÁVEL PARA ARMAZENAR O INPUT NOME
input_nome = tk.Entry(formulario, textvariable=var_nome, width=20)  # CRIA UM INPUT (CAMPO DE INSERÇÃO) E ARMAZENA O RESULTADO NA VARIÁVEL 'var_nome'
input_nome.pack(pady=3)     # COLOCA O INPUT NA TELA

label_idade = tk.Label(formulario, text="Insira sua idade:", font=('Arial', 10), bg='#c9c9c9')      # MESMA COISA DO NOME, MAS DESSA VEZ PARA A IDADE
label_idade.pack(pady=10)

var_idade = tk.IntVar()
input_idade = tk.Entry(formulario, textvariable=var_idade, width=3)
input_idade.pack(pady=3)

send_button = tk.Button(formulario, text='Enviar', command=get_info)    # CRIA UM BOTÃO PARA ENVIAR AS RESPOSTAS. A FUNÇÃO 'get_info()' É CHAMADO QUANDO O BOTÃO É CLICADO
send_button.pack(pady=10)   # COLOCA O BOTÃO NA TELA

janela_formulario.mainloop()    # MANTÉM A JANELA ABERTA ATÉ QUE SEJA FECHADA

print(f'Nome: {nome}\nIdade: {idade}')
```

MÉTODOS NOVOS:

- Label(): É usado para criar um widget de rótulo (label) em uma janela Tkinter. Um widget de rótulo é simplesmente um texto não interativo que é usado para exibir informações ou instruções para o usuário. Aqui estão os principais aspectos desse método:

    Sintaxe:

            widget_label = tk.Label(parent, options)

        - parent: Este é o widget pai onde o rótulo será colocado. Pode ser a janela principal ou outro widget, como um frame;
        - options: É um conjunto de opções que você pode fornecer para personalizar o rótulo. Alguns dos principais atributos que você pode definir são:

            - text: O texto que será exibido no rótulo;
            - font: A fonte usada para exibir o texto;
            - bg: A cor de fundo do rótulo;
            - fg: A cor do texto do rótulo;
            - width: A largura do rótulo em caracteres;
            - height: A altura do rótulo em caracteres;
            - anchor: A âncora que define como o texto é alinhado dentro do rótulo (por exemplo, 'center', 'w' para esquerda, 'e' para direita).

- Frame(): É usado para criar um widget de quadro (frame) em uma janela Tkinter. Um quadro é um recipiente retangular que pode conter outros widgets, como botões, campos de entrada, rótulos, etc. Aqui estão os principais aspectos desse método:

    Sintaxe: 

        frame = tk.Frame(parent, options)
        
    - parent: Este é o widget pai onde o quadro será colocado. Pode ser a janela principal ou outro quadro (frame) existente.

    - options: É um conjunto de opções que você pode fornecer para personalizar o quadro. Algumas das principais opções que você pode definir são:

        - bg: A cor de fundo do quadro;
        - bd: A largura da borda do quadro;
        - relief: O estilo da borda do quadro (por exemplo, 'flat', 'raised', 'sunken');
        - width: A largura do quadro em pixels;
        - height: A altura do quadro em pixels;

- grid(): Insere um widget e define sua geometria. Parâmetros:

    - row: A linha na qual o widget deve ser colocado (começando em 0);
    - column: A coluna na qual o widget deve ser colocado (começando em 0);
    - rowspan: O número de linhas que o widget deve ocupar;
    - columnspan: O número de colunas que o widget deve ocupar;
    - sticky: Define como o widget se comporta quando a célula é maior que o widget; pode ser uma combinação de "n", "s", "e", "w" (norte, sul, leste, oeste) para indicar o lado que deve grudar;
    - padx: Espaçamento horizontal em pixels em torno do widget;
    - pady: Espaçamento vertical em pixels em torno do widget.

- Configure(): É usado para modificar atributos ou opções de widgets. Você precisa fornecer o nome da opção que deseja configurar e o valor que deseja atribuir a essa opção.

- Variáveis de controle: Variáveis de controle em Tkinter são objetos especiais que servem como uma espécie de "ponte" entre widgets da interface gráfica (GUI) e seu programa Python. Elas são usadas para armazenar e rastrear valores associados a widgets, como campos de inserção (Entry), caixas de seleção (Checkbutton), botões de opção (Radiobutton), entre outros. Tkinter oferece vários tipos de variáveis de controle, dependendo do tipo de dado que você deseja armazenar. Os tipos mais comuns são:

    - StringVar(): Usado para armazenar strings;
    - IntVar(): Usado para armazenar valores inteiros;
    - DoubleVar(): Usado para armazenar valores de ponto flutuante;
    - BooleanVar(): Usado para armazenar valores booleanos (verdadeiro ou falso).

- Entry(): É usado para criar um widget de campo de inserção, onde os usuários podem inserir e editar texto. Esse widget é útil para receber informações de entrada do usuário, como nomes, senhas, endereços de e-mail, números de telefone, entre outros. Aqui estão os principais detalhes:

    Sintaxe: 

        widget_entry = tk.Entry(parent, options...)

    - parent: É o widget pai onde o campo de inserção será colocado. Normalmente, é a janela principal ou um quadro (Frame) que contém o campo de inserção.

    - options:

        - textvariable: Permite associar uma variável de controle (StringVar()) ao campo de inserção. Isso facilita a obtenção e a configuração do texto inserido no campo;
        - width: Define a largura do campo de inserção em caracteres;
        - show: Pode ser usado para ocultar o texto inserido, geralmente usado para senhas (ex: show='*').

- get(): É usado para recuperar o valor atual de uma variável de controle (como StringVar(), IntVar(), etc.) associada a um widget Tkinter. Ele é frequentemente usado para obter o texto inserido em um campo de inserção (Entry) ou o valor selecionado em um botão de opção (Radiobutton).

- insert(): É usado para inserir texto em um widget de entrada, como um campo de inserção (Entry). Ele permite adicionar texto em uma posição específica dentro do campo de inserção.

    Sintaxe: 

        widget_de_insercao.insert(index, text)

    - index: A posição onde o texto será inserido, geralmente especificada como um número inteiro que representa a posição do caractere (começando em 0);
    - text: O texto que será inserido no campo de inserção.

- delete(): É usado para remover texto de um widget de entrada, como um campo de inserção (Entry). Ele permite excluir uma parte específica do texto dentro do campo de inserção.

    Sintaxe: 
    
        widget_de_insercao.delete(start, end)

    - start: A posição de início da exclusão, geralmente especificada como um número inteiro que representa a posição do caractere (começando em 0).
    - end: A posição de término da exclusão (não inclusiva), também especificada como um número inteiro.

- quit(): É usado para encerrar a aplicação Tkinter. Ele fecha a janela principal e encerra o loop principal da interface gráfica, permitindo que o programa seja finalizado.
